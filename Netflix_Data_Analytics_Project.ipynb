# Data Analytics Internship Project
## Netflix Data Analysis

For this project, I will be analyzing the Netflix dataset to find some interesting trends. I will use pandas, matplotlib and seaborn for Data Analysis and Visualization.


```python
# Importing required libraries for analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

### 1. Loading Data
First, I will load the dataset using pandas read_csv method.


```python
# load the csv file
df = pd.read_csv('netflix_titles.csv')

# printing first 5 rows to see how data looks
df.head()
```

### 2. Checking Data Info
Let's check the shape and columns of the dataset to understand it better.


```python
# check total rows and columns
print("Shape:", df.shape)

# print all column names
print("Columns:", df.columns)
```

### 3. Missing Values
Now I will check if there are any null values in our dataset which needs cleaning.


```python
# finding missing values
missing_vals = df.isnull().sum()
print(missing_vals[missing_vals > 0])
```

### 4. Summary Statistics
Let's use the describe function to see some basic stats about the data.


```python
# using describe to get stats
df.describe(include='all')
```

### 5. Analyzing Patterns
I want to see the difference between the number of movies and TV shows, and also which country produces the most content.


```python
# check count of movies and tv shows
type_counts = df['type'].value_counts()
print(type_counts)

print("\n")
# top 5 countries
top_countries = df['country'].value_counts().head(5)
print(top_countries)
```

### 6. Data Visualization
Here I will create some graphs using matplotlib and seaborn to visualize the data properly.


#### 1. Checking Missing Data with Heatmap
First, a heatmap to see visually where the data is missing.


```python
plt.figure(figsize=(10,6))
# plot heatmap for null values
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

#### 2. Movies vs TV Shows Pie Chart
Let's visualize the distribution of content types.


```python
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Movies vs TV Shows')
plt.show()
```

#### 3. Top 10 Countries Bar Chart
Plotting a bar chart to see the top countries producing netflix content.


```python
plt.figure(figsize=(10,5))
top_10 = df['country'].value_counts().head(10)
sns.barplot(x=top_10.values, y=top_10.index, palette='Blues_r')
plt.title('Top 10 Countries')
plt.xlabel('Count')
plt.ylabel('Country Name')
plt.show()
```

#### 4. Release Year Histogram
Let's see the trend of content added over the years. I will only consider data after year 2000 for better graph.


```python
plt.figure(figsize=(10,5))
df_recent = df[df['release_year'] >= 2000]
sns.histplot(df_recent['release_year'], bins=20, color='purple', kde=True)
plt.title('Content Release By Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

### Conclusion
Based on my analysis of the Netflix dataset, I found out following things:
1. Netflix has lot more Movies (around 70%) compared to TV shows.
2. Most of the content on Netflix comes from the United States, and India is on second position.
3. There is a lot of missing data in director and cast columns which I saw using the heatmap.
4. From the histogram, we can see that Netflix started adding a lot of content rapidly after 2015.
